Create a script to develop a Python function for descriptive statistics. The input for the function should be the sample and the field to perform the descriptive statistics

```python
import pandas as pd
import numpy as np
df = pd.read_csv("daily_food_delivery_orders.csv")

def descriptive_stats(sample, field):
    
    data = sample[field]
    
    stats = {
        "Count": data.count(),
        "Mean": data.mean(),
        "Median": data.median(),
        "Mode": data.mode()[0],
        "Minimum": data.min(),
        "Maximum": data.max(),
        "Range": data.max() - data.min(),
        "Variance": data.var(),
        "Standard Deviation": data.std(),
        "Q1": data.quantile(0.25),
        "Q2 (Median)": data.quantile(0.50),
        "Q3": data.quantile(0.75),
        "IQR": data.quantile(0.75) - data.quantile(0.25),
        "Skewness": data.skew()
    }
    
    return stats
descriptive_stats(df, "delivery_time_minutes")
```

Visualize the dependent variable by the Graph/Chart of the following using Python Program: Heatmap

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("daily_food_delivery_orders.csv")

pivot_table = pd.pivot_table(df, index='restaurant_type', columns='payment_method', values='delivery_time_minutes', aggfunc='sum')

plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, fmt=".1f", cmap='coolwarm', cbar_kws={'label': 'Total delivery timem'})

plt.ylabel('Restaurant Type')
plt.xlabel('Payment Method')
plt.title('Heatmap of Total delivery time by Restaurant Type and Payment Method')

plt.show()
```

Hypothesis Test(Correlation)

```python
import pandas as pd
from scipy.stats import pearsonr

df = pd.read_csv("daily_food_delivery_orders.csv")

df_clean = df.dropna(subset=['delivery_distance_km', 'delivery_time_minutes'])

r, _= pearsonr(df_clean['delivery_distance_km'], df_clean['delivery_time_minutes'])

print(f"Pearsons's Correlation Coefficient: {r:.4f}")
```

One Sample T-test

```python
import pandas as pd
from scipy.stats import ttest_1samp

df = pd.read_csv("daily_food_delivery_orders.csv")

t_stat, p_value = ttest_1samp(df['delivery_time_minutes'],500)

print("T-statistic: ",t_stat)
print("P-value: ",p_value)
```
